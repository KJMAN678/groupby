```python
import pandas as pd
import os
import zipfile
import requests
import io
from IPython.display import display

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 500)
```

```python
path = os.getcwd()
```

```python
# データがあるurlの指定
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip'

# データをurlから取得
r = requests.get(url, stream=True)

# zipfile内のstudent-mat.csvをカレントディレクトリ内のdataディレクトリに展開
with zipfile.ZipFile(io.BytesIO(r.content)) as existing_zip:
    existing_zip.extract('student-mat.csv', './')
```

```python
df = pd.read_csv(path + "\student-mat.csv", sep = ';')
```

```python
display(df.head())
display(df.tail())
display(df.dtypes)
```

```python
df.groupby('sex').agg(
    count=('sex', 'count'),
    age_mean=('age', 'mean'),
    G1_mean=('G1', 'mean'),
    G1_median=('G1', 'median'),
    G2_mean=('G2', 'mean'),
    G2_median=('G2', 'median'),
    G3_mean=('G3', 'mean'),
    G3_median=('G3', 'median')
)
```
